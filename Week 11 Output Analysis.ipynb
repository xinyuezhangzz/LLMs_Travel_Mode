#### (3) DF1 & DF2

```python
def calculate_rates(df):
    # Correctly filtering out both 'na' and 'NA'
    extraction_rate = len(df[~df['Travel Mode'].isin(['na', 'NA'])]) / len(df)
    satisfaction_rate = len(df[df['Sentiment'] == 'positive']) / len(df[~df['Travel Mode'].isin(['na', 'NA','NaN'])])
    self_verification_rate = len(df[df['Self Verification Result'] == 'correct']) / len(df)
    
    # Check if 'Cross Verification Result' exists in the DataFrame
    if 'Cross Verification Result' in df.columns:
        cross_verification_rate = len(df[df['Cross Verification Result'] == 'correct']) / len(df)
    else:
        cross_verification_rate = None  # Use None for DataFrames without cross-verification data

    return extraction_rate, satisfaction_rate, self_verification_rate, cross_verification_rate
```

```python
# Calculate rates for each DataFrame
rates_df1 = calculate_rates(df1)
rates_df2 = calculate_rates(df2)
rates_df1_base = calculate_rates(df1_base)
rates_df2_base = calculate_rates(df2_base)

# Creating a list for each DataFrame rates to include in the summary
data_rates = [rates_df1_base, rates_df2_base, rates_df1, rates_df2]

# Creating the summary DataFrame
summary_df = pd.DataFrame(data_rates, columns=['Extraction Rate','Satisfaction Rate','Self-Verification Rate', 'Cross-Verification Rate'])
summary_df.index = ['baseline df1', 'baseline df2','final experiment df1', 'final experiment df2']

# Displaying the summary DataFrame
summary_df
```

### 4. Data Visualization

#### (1) DF1: 60900_61899

```python
explode_travel_mode1 = [0.1 if mode == travel_mode_counts1.idxmax() else 0 for mode in travel_mode_counts1.index]

# Plot for Travel Mode Distribution with explode for the largest piece
plt.figure(figsize=(8, 6))
plt.pie(travel_mode_counts1, labels=travel_mode_counts1.index, autopct='%1.1f%%', startangle=140, explode=explode_travel_mode1)
plt.title('Distribution of Travel Modes (Tweet 60900_61899)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.legend(travel_mode_counts1.index, title="Travel Mode", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.tight_layout()  # Adjusts plot parameters for a tight fit
plt.show()
```

```python
# count uber,lyft as taxi
# Copying the 'Travel Mode' series to avoid changing df1
travel_mode_series = df1['Travel Mode'].replace(['uber', 'lyft'], 'taxi').copy()

# Now, calculate the counts for the modified series
travel_mode_counts1_mod = travel_mode_series.value_counts()

# Generate the explode list for the modified counts
explode_travel_mode1_mod = [0.1 if mode == travel_mode_counts1_mod.idxmax() else 0 for mode in travel_mode_counts1_mod.index]

# Plot for Travel Mode Distribution with modified counts
plt.figure(figsize=(8, 6))
plt.pie(travel_mode_counts1_mod, labels=travel_mode_counts1_mod.index, autopct='%1.1f%%', startangle=140, explode=explode_travel_mode1_mod)
plt.title('Distribution of Travel Modes-taxi adjusted (Tweet 60900_61899)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.legend(travel_mode_counts1_mod.index, title="Travel Mode", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.tight_layout()  # Adjusts plot parameters for a tight fit
plt.show()
```

```python
explode_sentiment1 = [0.1 if sentiment == Sentiment_counts1.idxmax() else 0 for sentiment in Sentiment_counts1.index]
# Plot for Sentiment Distribution with explode for the largest piece
plt.figure(figsize=(8, 6))
plt.pie(Sentiment_counts1, labels=Sentiment_counts1.index, autopct='%1.1f%%', startangle=140, explode=explode_sentiment1)
plt.title('Distribution of Sentiments (Tweet 60900_61899)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.legend(Sentiment_counts1.index, title="Sentiment", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.tight_layout()  # Adjusts plot parameters for a tight fit
plt.show()
```

```python
# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(contingency_table1, annot=True, fmt="d", cmap="YlGnBu", cbar=True)
plt.title('Heatmap of Travel Mode vs Sentiment (Tweet 60900_61899)')
plt.xlabel('Sentiment')
plt.ylabel('Travel Mode')
plt.show()
```

```python
# Plot a stacked bar chart of proportions
proportion_table1.plot(kind='bar', stacked=True, colormap='viridis', figsize=(10, 7))
plt.title('Proportion of Sentiment Within Each Travel Mode (Tweet 60900_61899)')
plt.xlabel('Travel Mode')
plt.ylabel('Proportion')
plt.legend(title='Sentiment', loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```
